```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import missingno as msno
import plotly.express as px
```

```python
dataset=pd.read_csv('/content/netflix_titles.csv')
dataset.head()
```

```python
#basic details of the data
print("Shape of data:", dataset.shape)
```

```python
#datatypes info
dataset.info()
print("Summary:\n")
dataset.describe(include='all')
```

```python
#missing values in each columns
dataset.isnull().sum()
```

```python
#missing values
msno.matrix(dataset)
```

```python
#top countries according to the netflix content
top_countries = dataset['country'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
plt.title('Top 10 Countries with the Most Movies/TV Shows on Netflix')
plt.xlabel('Number of titles')
plt.ylabel('country')
plt.show()
```

```python
sns.countplot(data=dataset, x='type', palette='Set2')
plt.title("Count of movies vs TV shows on Netflix")
plt.show()
```

```python
#yearwise content change
dataset['date_added']=dataset['date_added'].astype(str).str.strip()
dataset['date_added']= pd.to_datetime(dataset['date_added'])
dataset['year_added'] = dataset['date_added'].dt.year
plt.figure(figsize=(10,6))
sns.histplot(data=dataset, x='year_added', bins=20, kde=True)
plt.title("Netflix Content Added Over the Years")
plt.show()
```

```python
#correlation heatmap
dataset_encoded = pd.get_dummies(dataset[['type', 'rating']], drop_first=True)
plt.figure(figsize=(10,6))
sns.heatmap(dataset_encoded.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#most frequent genres data
genres = dataset['listed_in'].str.split(', ', expand=True).stack().value_counts().head(10)
px.bar(x=genres.index, y=genres.values, labels={'x':'Genre', 'y':'Count'}, title='Top 10 Netflix Genres')
```

```python
#movies distribution according to their duration
movies= dataset[dataset['type']=='Movie'].copy()
movies['duration_mins'] = movies['duration'].str.replace(' min', '').replace('No data', np.nan)
movies['duration_mins'] = pd.to_numeric(movies['duration_mins'], errors='coerce')
sns.boxplot(data=movies, y='duration_mins', color='skyblue')
plt.title("Distribution of movie duration")
plt.show()
```
